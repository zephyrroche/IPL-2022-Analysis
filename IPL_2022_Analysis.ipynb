# **What is the venue with the most matches hosted in IPL 2022?**

```python
venue_counts = data['venue'].value_counts()
venue_counts
```

```python
venue_counts = data.groupby('venue')['year'].sum().reset_index()
```

```python
fig = px.pie(venue_counts,
             values='year',
             names='venue',
             title='Top 5 IPL Venues',
             hole=0.6, color_discrete_sequence=px.colors.qualitative.Pastel)
fig.show()
```

# **Did winning the toss often lead to winning the match in IPL 2022?**

```python
data.head(1)
```

```python
# Create a column to indicate if toss winner also won the match
data['toss_win_match_win'] = data['toss_winner'] == data['match_winner']

# Calculate percentage of matches where toss winner won
toss_success_rate = data['toss_win_match_win'].mean() * 100
print(f"Toss winner also won the match in {toss_success_rate:.2f}% of IPL 2022 games.")
```

# **What is the win percentage for each team in IPL 2022?**

```python
# Total matches played per team
total_matches = pd.concat([data['team1'], data['team2']]).value_counts()

# Matches won per team
wins = data['match_winner'].value_counts()

# Calculate win percentage
win_percentage = (wins / total_matches * 100).sort_values(ascending=False)
print(win_percentage)
```

```python
win_percentage.plot(kind='barh', title='Win Percentage per Team - IPL 2022', xlabel='Win %', color='purple')
```
