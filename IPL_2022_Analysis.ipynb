```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('IPL.csv')
```

# **Number of matches won by each team in IPL 2022**

```python
figure = px.bar(data, x= data['match_winner'], title = 'Number of Matches won in IPL 2022')

figure.show()
```

Gujarat won the most

```python
data['won_by'] = data['won_by'].map({'Wickets':'Chasing',
                                     'Runs':'Defending'})
won_by = data['won_by'].value_counts()
label = won_by.index
counts = won_by.values

colors = ['red','lightgreen']

fig = go.Figure(data=[go.Pie(labels = label, values = counts)])
fig.update_layout(title_text = 'Number of matches won by defending or chasing')
fig.update_traces(hoverinfo = 'label+percent', textinfo = 'value', textfont_size = 30,
                  marker = dict(colors = colors, line = dict(color = 'black', width=3)))
```

```python
figure = px.bar(data, x=data['best_bowling'], title='Best Bowler in IPL 2022')
figure.show()
```

Yuzvendra Chahal is the best bowler in the IPL 2022

```python
figure = px.bar(data, x=['player_of_the_match'],
                title='Most Player of the match awards')
figure.show()
```

Kuldeep Yadav won the most Player of the match awards

# **Top Scorer of IPL 2022**

```python
figure = px.bar(data, x=data['top_scorer'],
                y=data['highscore'],
                color=data['highscore'], title='Top Scorers in IPL 2022')
figure.show()
```

Quinton de Knock has so far scored the highest out of everyone (140) although Jos Buttler has a larger amount of scores.
